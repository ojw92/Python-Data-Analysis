```python
# 5 - 2    visualizing data

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
        
plt.title('Highest Temperature')
plt.plot(result, 'r')
plt.show()
```

```python
# filtering just the August data

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':    # row = ['1970-10-01', '108', '13.5', '7.9', '20.7']
            result.append(float(row[-1]))

# plt.figure(figsize=(10,2))
plt.plot(result, 'hotpink')
plt.show()
f.close()
```

```python
# filtering just February 14

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
            result.append(float(row[-1]))
          
plt.figure(figsize=(10,2))
plt.plot(result, 'deeppink')
plt.show()
```

```python
# highest & lowest temperature data after 1983

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        if 1983 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
plt.figure(figsize = (10, 2))
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')
plt.show()
f.close()
```
