```python
# 6 - 3    histogram of highest temperature data (1907~2021)
# recall that histogram shows frequency of data

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.hist(result, bins=100, color='r')
plt.show()
```

```python
# only August

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '08':
            result.append(float(row[-1]))
            
plt.hist(result, bins=100, color='r')
plt.show()
f.close()
```

```python
# January & August data

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))
            
plt.figure(figsize = (10,5))
plt.title('January & August Highest Temperatures')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.hist(aug, bins=100, color='r', label='Aug')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
f.close()
```

```python
# 6 - 4    boxplot

import matplotlib.pyplot as plt
import random
result = []
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))    # order by size

plt.boxplot(result)
plt.show()
```

```python
# boxplot shows maximum & minimum as whiskers
# median as orange line in the middle
# 25th percentile and 75th percentile (Q1 & Q3) as the box's bases
# 25th percentile means 25% of the data is below it
```

```python
# to find out the exact values of the 25th, 50th, and 75th percentile

import numpy as np
result = np.array(result)
print("1/4: "+str(np.percentile(result,25)))
print("2/4: "+str(np.percentile(result,50)))
print("3/4: "+str(np.percentile(result,75)))
```

```python
# boxplot for each and every month

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]    # list of 12 lists, 1 for each month

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))    # n-th month data stored in n-th list

plt.boxplot(month)
plt.show()
```

```python
# in the above, January data is stored in month[0], the first list
# February data stored in month[1], the 2nd list, and so on
```

```python
# boxplot for days of august

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):    # add 31 lists, for 31 days
    day.append([])

for row in data:
    if row[-1] != '':
        date = row[0].split('-')
        if date[1] == '08':
            day[int(date[2])-1].append(float(row[-1]))

plt.style.use('ggplot')    # plot style - gray background + blue median
plt.figure(figsize=(10,5), dpi=300)
plt.title('Highest Temperatures for Days of August')
plt.boxplot(day, showfliers=False)    # remove outliers
plt.xlabel('Days')
plt.ylabel('Temperatures (C)')
plt.show()
f.close()
```

```python
# Daily temp diff () for each month

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
month =[]

for i in range(12):
    month.append([])
    
for row in data:
    if row[-1] != '' and row[-2] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1])-float(row[-2]))
        
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title('Daily Temperature Difference () for Each Month')
plt.boxplot(month, showfliers=False)
plt.xlabel('Months')
plt.ylabel('Temperature Differences (C)')
plt.show()
f.close()
```
