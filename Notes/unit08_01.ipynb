```python
# taking input to sector and graphing its age data

import csv
import matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',', '')))    # replace commas in string with '' (ex. 5,000 --> 5000)
            
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(name+' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
# bar graph on sector age data

import csv
import matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)
result = []

for row in data:
    if '북아현' in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))
            
plt.bar(range(101), result)    # 0 ~ 100+ age groups = 101
plt.show()
f.close()
```

```python
# 8 - 2

import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '북아현' in row[0]:
        for i in range(0,101):
            m.append(int(row[3+i]))    # male data from row[3] to row[103]
            f.append(int(row[-1-i]))    # female data from row[-1] to row[-101] (reverse order, age 100+ ~ 0)
f.reverse()    # reverse the female data to 0 ~ 100+
```

```python
# visualizing data using the 2nd (non-reversing) method

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '북아현' in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for i in row[106:207]:
            f.append(int(i))
            
plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()
```

```python
# this causes the two horizontal bar graphs to overlap
# make male data negative:

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '북아현' in row[0]:
        for i in row[3:104]:
            m.append(-int(i))    # male data on left (negative)
        for i in row[106:]:
            f.append(int(i))
            
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False    # need this since we have negative data
plt.title('북아현 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
# full male & female age data represented as horizontal bar graph, with user input of sector name

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]:
        for i in range(0,101):
            m.append(-int(row[i + 3]))
            f.append(int(row[-(i + 1)]))
f.reverse()

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+' 지역의 남녀 성별 인구 분포')
plt.barh(range(0,101), m, label='남성')
plt.barh(range(0,101), f, label='여성')
plt.xlabel('Number of People')
plt.ylabel('Age')
plt.legend()
plt.show()
```
