```python
# recall the 항아리 모양 그래프

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '북아현' in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.barh(range(0,101),m)
plt.barh(range(0,101),f)
plt.show()
```

```python
# 9 - 1    Jeju

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i.replace(',','')))
        for i in row[106:]:
            f.append(int(i.replace(',','')))

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(10,5), dpi=300)
plt.title(name+' 지역의 남녀 성별 인구 분포')
plt.barh(range(0,101), m, label='남성')
plt.barh(range(0,101), f, label='여성')
plt.xlabel('Number of people')
plt.ylabel('Age')
plt.legend()
plt.show()
```

```python
# ValueError - x-values (age groups 0 ~ 100+) do not match data found
# there're 46 rows with 제주특별자치도 in row[0] --> 101 * 46 = 4646 data
# we only want that single row that represents the whole 제주도
```

```python
# method we'll use: stop searching for more rows of 제주특별자치도 once we find the 1st row with 제주특별자치도
# break the search loop using 'break'

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i.replace(',','')))
        for i in row[106:]:
            f.append(int(i.replace(',','')))
        break

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+' 지역의 남녀 성별 인구 분포')
plt.barh(range(0,101), m, label='남성')
plt.barh(range(0,101), f, label='여성')
plt.xlabel('Number of people')
plt.ylabel('Age')
plt.legend()
plt.show()
```

```python
# if the pie chart looks like an ellipse and not a circle, use axis()

import matplotlib.pyplot as plt
size = [2441, 2312, 1031, 1233]
plt.axis('equal')
plt.pie(size)
plt.show()
```

```python
# adding labels to each slice

import matplotlib.pyplot as plt
size = [2441, 2312, 1031, 1233]
label = ['A형', 'B형', 'AB형', 'O형']
plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
plt.pie(size, labels=label)
plt.legend()
plt.show()
```

```python
# notice that pie chart starts at 3 o'clock and fills up in counter clock-wise direction
```

```python
# to include percentages, use 'autopct' (auto percent) 

import matplotlib.pyplot as plt
size = [2441, 2312, 1031, 1233]
label = ['A형', 'B형', 'AB형', 'O형']
plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%')    # percentage rounded to nearest 10ths place
plt.legend()
plt.show()
```

```python
# changing color and using explode feature

import matplotlib.pyplot as plt
size = [2441, 2312, 1031, 1233]
label = ['A형', 'B형', 'AB형', 'O형']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0, 0, 0.1, 0))    # explode 3rd entry
plt.legend()
plt.show()
```

```python
# pie chart for Jeju male & female populations

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3].replace(',',''))
            f += int(row[-(i+1)].replace(',',''))    # or row[i+106]
        break
size.append(m)
size.append(f)
print(size)

plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
color = ['crimson', 'darkcyan']
plt.title(name+' 지역의 남녀 성별 비율')
plt.pie(size, labels=['남','여'], autopct='%.2f%%', colors=color, startangle=90)
plt.legend()
plt.show()
```

```python
# startangle = 90 causes the pie chart to start at 90 degrees from 3 o'clock
```
