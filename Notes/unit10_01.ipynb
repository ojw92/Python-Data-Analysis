```python
# 10 - 1    꺾은선 그래프

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요 : ')

for row in data:
    if name in row[0]:
        for i in range(3, 104):    # same count as range(101)
            m.append(int(row[i].replace(',','')))    # instead of i + 3
            f.append(int(row[i+103].replace(',','')))    # instead of i + 106
        break

plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

```python
# 10 - 2 bar graph (difference between male & female population)

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
result = []
name = input('궁금한 지역을 입력하세요 : ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
        break
            
print(len(result))
plt.bar(range(101), result)
plt.show()
```

```python
# 100 random points

import random
import matplotlib.pyplot as plt
x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
plt.scatter(x, y, size, c=size, cmap='jet')    # colored by size
plt.colorbar()
plt.show()
```

```python
# Jeju population data using colored scatterplot

import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요: ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break

plt.style.use('ggplot')
plt.scatter(m, f, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```

```python
# apply size variation by taking the combined male & female population and taking square root

import csv
import math
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input('궁금한 동네를 입력해주세요 : ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
            size.append(math.sqrt(int(row[i].replace(',',''))+int(row[i+103].replace(',',''))))    # sqrt of sum
        break    # only search for first 동네 match (제주도 전체)

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m, f, s=size, c=range(101), alpha=0.7, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
# f.close()    # f taken up by f=[]
```
