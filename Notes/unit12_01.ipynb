```python
# 12 - 4    most boarding station for each hour (0~23)

import csv
import matplotlib.pyplot as plt
f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24    # max station for each of 24 hour
mx_station = [''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):    # find and record 24 hours from row
        a = row[j*2 + 4]    # index of boarding passengers for each hour is represented by i = 2*j + 4
        if a > mx[j]:
            mx[j] = a
            if j+4 > 24:    # last 3 hrs are 25, 26, 27 hr, so correct them to 1, 2, 3
                j = j - 24
            mx_station[j] = row[3]+' ('+str(j+4)+'시)'    # station & hour
print(mx_station)
print(mx)

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)    # add label for each x-axis tick and rotate by 90 degrees
plt.show()
```

```python
# 신림 7 times, 강남 10 times --> 강남 is the most popular station where people board
# 출근 시간대 = 신림 has the most boarding
# afternoon ~ midnight = 강남 has the most boarding
```

```python
# most exiting station for each hour

import csv
import matplotlib.pyplot as plt
f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        b = row[2*j+5]
        if b > mx[j]:
            mx[j] = b
            mx_station[j] = row[3]+' ('+str(j+4)+'시)'
print(mx_station, mx)

plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.bar(range(24), mx, color='b')
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
# 출근 시간대 ~ afternoon = 강남 has the most exits
# afternoon ~ midnight = 신림 has the most exits
```

```python
# 12 - 5    total board & exit for all stations for each hour

import csv
f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
s_in = [0] * 24    # total boarded for all stations, for each hr
s_out = [0] * 24    # for exited

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4 + i*2]    # insert boarding data by hr, summing with previous row's entries
        s_out[i] += row[5 + i*2]

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, 'r', label='승차')    # no need to enter range(24) for x-values ...
plt.plot(s_out, 'g', label='하차')
plt.xlabel('시간대')
plt.ylabel('인원수')
plt.legend()
plt.xticks(range(24), range(4,28))    # ... because this takes care of it
plt.show()
```

```python
# 출근 시간대: most boarded during 7~9 am & 5~7pm
# 퇴근 시간대: most exited during 7~9am & 5~7 pm
# notice the 1e7 above the y-axis (magnitude of 10^7)
```
