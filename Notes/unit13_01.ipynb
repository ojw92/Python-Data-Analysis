```python
# 13 - 1    numpy tutorial

import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)    # evenly sampled time at 200 ms intervals

plt.plot(t, t, 'r--',   t, t**2, 'bs',   t, t**3, 'g^')    # red dashes, blue squares, green triangles
plt.show()
```

```python
# 13 - 3    graphing with numpy

import matplotlib.pyplot as plt

# histogram graphed with unit 6's random library
import random
dice = []
for i in range(10):
    dice.append(random.randint(1,6))

plt.hist(dice, bins=6)
plt.show()

# histogram graphed with numpy
import numpy as np
dice = np.random.choice(6, 10)    # much simpler

plt.hist(dice, bins=6)
plt.show()
```

```python
# when doing 100,000+ iterations, numpy is faster
```

```python
# storing numbers in a sequence or in a pattern

import numpy as np
print(np.arange(3))    # 0 to 3-1
print(np.arange(3,7))    # 3 to 7-1
print(np.arange(3,7,2))    # 3 to 7-1 in intervals of 2

a = np.arange(1, 2, 0.1)    # choose value of intervals
b = np.linspace(1, 2, 11)    # choose number of values; include last number
print(a)
print(b)

a = np.arange(-np.pi, np.pi, np.pi/10)    # divide whole range by 20 and exclude last number
b = np.linspace(-np.pi, np.pi, 20)    # divide the whole range by 20-1 and include last number
print(a)
print(b)
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))    # lateral shift due to changing domain
plt.show()
```

```python
# masking - applying conditions to array to pick out the values you want
# take notice that a mask has holes in specific places, as needed for its purpose

a = np.arange(-5, 5)
print(a)
print(a < 0)    # masking
print(a[a < 0])    # putting mask back on original array, we can filter desired values

mask1 = abs(a) > 3
print(a[mask1])

mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1 + mask2])    # OR (either has to be True)
print(a[mask1 * mask2])    # AND (both have to be True)
```

```python
# applying masks on bubble chart

import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000)    # 1000 values
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(100) * 100    # array of 100 numbers between 0 and 1, times 100
mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 + mask2]
y = y[mask1 + mask2]
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# masks can be used to clear the clutter and filter out the values that require more attention
```
