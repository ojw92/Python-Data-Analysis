```python
# sort golden medals in ascending order

import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table',
                 header=0,
                 index_col=0)
summer = df[1].iloc[:,:5]
summer.columns = ['경기수', '금', '은', '동', '계']
print(summer.sort_values('금', ascending=False))    # to sort by ascending, set to True
summer.sort_values('금', ascending=False)    # visualize as table
```

```python
# make mask to filter data that meets certain criteria

import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
print(df['B'] > 0.4)
```

```python
# mathematic operations on DataFrame involves calculation in row direction or column direction
# axis = 0 is row direction
# axis = 1 is column direction
# '행렬' <-- 행 before 열

# example of adding in row direction

dftest = pd.DataFrame(np.array(([1,2,3], [3,4,5], [5,6,7])), 
                    index = ['row 1', 'row 2', 'row 3'], 
                     columns = list('ABC'))    # add 3 rows in sequence
print(dftest)
dftest['D'] = np.sum(dftest, axis=1)    # create a new column by sum in row direction
dftest
```

```python
import pandas as pd
df = pd.read_csv('age.csv', encoding='cp949', index_col=0)
df = df.div(df['총인구수'], axis=0)    # represent entire data as proportions by dividing by total population
del df['총인구수'], df['연령구간인구수']    # delete the 2 unneeded columns
```

```python
# make sure csv file has integers (일반) cells where they need to be
# if cell entries are 통화 (currency) type due to being large numbers with comma placeholders,
# they could be read as strings, and cannot be have math operations done on them
```

```python
# take input and store its demographic data

name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)    # find the index of the index column that contains the string
df2 = df[a]
df2
```

```python
# to plot, flip row & column via transpose first, then use plot()

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
df2.T.plot()
plt.show()
```

```python
# in df.plot(), rows become x-axis (each row is a value on x-axis)
#               values in each of the cells become y-values
#               colums become different graphs (each column becomes an individual plot)
# we need to know what we want as x-values, y-values and legends, and use df.T.plot() accordingly
```

```python
# visualize similar demographics

import numpy as np
x = df.sub(df2.iloc[0], axis=1)    # subtract first row of df2 input match result from all of df
y = np.power(x, 2)    # square of the difference (DataFrame)
z = y.sum(axis=1)    # sum up in row direction (1 column; 1 value for each and all neighborhoods)
```

```python
# because df2.iloc[0] is a row of values, we want to subtract those values in row direction from each of df's rows
# in short, subtracting a Series from DataFrame automatically subtracts from all arrays of the DataFrame

# row operations --> axis = 1
# column operations --> axis = 0
```

```python
# sort & slice to find top n number of results

i = z.sort_values().index[:5]    # top 5 minimal values in squared sum
df.loc[i].T.plot()
plt.show()
```

```python
# DataFrame uses .iloc[row:column]
# Series uses .loc[row]
```

```python
# combined code

x = df.sub(df.iloc[0], axis=1)
y = np.power(x, 2)
z = y.sum(axis=1)
i = z.sort_values().index[:5]
df.loc[i].T.plot()
```

```python
# Final version

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('age.csv', encoding='cp949', index_col=0)
df = df.div(df['총인구수'], axis=0)
del df['총인구수'], df['연령구간인구수']
df = df.fillna(0)    # changes NaN values to 0 in df

name = input('원하는 지역의 이름을 입력해 주세요 : ')
a = df.index.str.contains(name)
df2 = df[a]

plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
#plt.title(name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
p = df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot(figsize=(20,10),
                                                            title=name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
p.set_xlabel('연령대')
p.set_ylabel('인구 비율')
plt.show()
```
